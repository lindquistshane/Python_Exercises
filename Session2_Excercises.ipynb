```python
#insert a new column which contains the sum of the values in the other 'leaf' columns
cols = [
    "leafspots-halo",
    "leafspots-marg",
    "leafspot-size",
    "leaf-shread",
    "leaf-malf",
    "leaf-mild"
]

#the values need to be cast as numbers in order to sum, as we saw in the prior step they are currently 'objects'
df['leaf_disease'] = df[cols].apply(pd.to_numeric, errors="coerce").sum(axis=1)
df['leaf_disease'].head()
```

```python
#insert a new column called 'leaf_category' use it to categorize the values in leaf_disease to low:<2, medium:3-5, or high:>5
leaf_disease = df['leaf_disease']
bins = [0,2,5,100]
bin_names = ['low', 'medium', 'high']
df['leaf_category'] = pd.cut(leaf_disease,bins,labels=bin_names)
```

```python
#Store the new dataframe as a .csv file in the present working directory
df.to_csv('soybean2.csv', index=False)
```

```python
#Delete the two new columns we just created
df.drop(columns = ['leaf_disease', 'leaf_category'], inplace = True)
```

```python
#Insert a new column called "disease_average" in the soybean dataframe. The new column is the average of the values of the columns starting with "area-damaged" until the column (inclusive of) "sclerotia"
cols2 = [
    'area-damaged',
    'severity',
    'seed-tmt',
    'germination',
    'plant-growth',
    'leaves',
    'leafspots-halo',
    'leafspots-marg',
    'leafspot-size',
    'leaf-shread',
    'leaf-malf',
    'leaf-mild',
    'stem',
    'lodging',
    'stem-cankers',
    'canker-lesion',
    'fruiting-bodies',
    'external decay',
    'mycelium',
    'int-discolor',
    'sclerotia'
]

#the values need to be cast as numbers in order to sum, as we saw in the prior step they are currently 'objects'
df['disease_average'] = df[cols2].apply(pd.to_numeric, errors="coerce").mean(axis=1)
df['disease_average'].head()
```

```python
#Sort the dataframe in descending order of the values in the new column you created ("disease_average")
df2 = df.sort_values(by="disease_average")
df2
```

----------

```python
#insert a new column called "disease_category", the new column should have the categories low, medium, and high depending on the value of the disease_average column (<0.5, 0.5 - 1, >1)
disease_average = df['disease_average']
bins = [0,.5,1,100]
bin_names = ['low', 'medium', 'high']
df['disease_category'] = pd.cut(disease_average,bins,labels=bin_names)
df.loc[:,['disease_category','disease_average']]
```

```python
#count the number of records per the "disease_category" column
print(pd.Series(df['disease_category']).value_counts())
```

```python
#how many records in each of the disease categories are treated for seed disease (seed-tmt will be 1)
df.loc[df["seed-tmt"] == "1", "disease_category"].value_counts()
```

```python
#Create a pivot table to display the average value of "area-damaged" for each of the disease categories and seed treatment status ("seed-tmt")
import numpy as np
df["area-damaged"] = pd.to_numeric(df["area-damaged"], errors="coerce")
pivotdf = pd.pivot_table(df, values = 'area-damaged',
                         index=['disease_category'], columns=['seed-tmt'], aggfunc= np.mean)
pivotdf
```
